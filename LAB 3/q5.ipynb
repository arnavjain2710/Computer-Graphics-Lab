```python
import numpy as np
import matplotlib.pyplot as plt

# Create array from 0 to 2
x = np.linspace(0, 2 * np.pi, 100)

y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, label='y = sin(x)', color='blue', marker='o')
plt.plot(x, y_cos, label='y = cos(x)', color='red', marker='x')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = sin(x) and y = cos(x)')
plt.legend()

plt.show()
```
