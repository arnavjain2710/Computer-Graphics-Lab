```python
import matplotlib.pyplot as plt
def draw_circle(xc, yc, r , print_points = True):
    x, y = 0, r  
    p = 1 - r    
    points = []  

    def plot_circle_points(xc, yc, x, y):
        points.extend([(xc + x, yc + y), (xc - x, yc + y),
                       (xc + x, yc - y), (xc - x, yc - y),
                       (xc + y, yc + x), (xc - y, yc + x),
                       (xc + y, yc - x), (xc - y, yc - x)])

    plot_circle_points(xc, yc, x, y)

    while x < y:
        if print_points:
            print(points)
        x = x + 1
        if p < 0:
            p += 2*x + 1  
        else:
            y -= 1
            p += 2*x - 2*y + 1  

        plot_circle_points(xc, yc, x, y)

    if print_points:
        print(points)

    x_cordinate, y_cordinate = zip(*points)
    plt.scatter(x_cordinate, y_cordinate, s=20)
    plt.gca().set_aspect('equal')
    plt.show()
```

```python
draw_circle(0, 0, 10 , False)
```

```python
draw_circle(2, 3, 5 , False)
```
